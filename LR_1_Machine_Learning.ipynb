У контексті машинного навчання та обробки зображень поняття “форма” (shape) використовується для опису розмірності масивів даних, які представляють як зображення, так і відповідні мітки.

Форма тренувальних зображень – це характеристика, яка вказує, якою є структура  (розмірність) даних, що представляють зображення, у вигляді багатовимірного масиву.

Форма описує, скільки зображень є в наборі, а також розміри кожного зображення  і кількість каналів кольорів. Наприклад, якщо в наборі 60 000 зображень, де кожне зображення має розмір 28x28 пікселів  і є чорно-білим (1 канал), то форма може бути представлена як (60000, 28, 28, 1)

Форма тренувальних міток (labels) вказує, як представлені дані, що відповідають класовим  міткам або значенням цільової змінної для кожного зображення.

Наприклад, якщо кожне з 60 000 зображень має відповідну мітку у вигляді цілого числа, то форма міток може бути такою (60000,)

```python
 # 3.3. Titanic

import pandas as pd
import seaborn as sns

# 3.3.2. Завантаження та ознайомлення з даними
# a. Завантажте набір даних за допомогою бібліотеки seaborn або pandas.
try:
    titanic = sns.load_dataset('titanic')
    print("Завантажено Titanic за допомогою seaborn.")
except:
    url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
    titanic = pd.read_csv(url)
    print("Завантажено Titanic за допомогою pandas з URL.")

# Print the columns of the DataFrame to identify the actual column names
print("\nColumns in the Titanic DataFrame:")
print(titanic.columns)

print(titanic.head())

# b. Перевірте розміри набору даних та інформацію про кожну колонку
# (типи даних, кількість пропущених значень).
print("Розмір набору даних (рядки, стовпці):", titanic.shape)
print("\nІнформація про колонки (типи даних, кількість не-null значень):")
print(titanic.info())
print("\nОпис числових колонок:")
print(titanic.describe())

# 3.3.3. Попередня обробка даних
# a. Обробіть пропущені значення:
# Заповніть пропущені значення у Age (наприклад, середнім або медіанним значенням).
if 'Age' in titanic.columns:  # Check if 'Age' column exists
    titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
    print("\nПропущені значення в Age заповнені медіаною.")
else:
    print("\nКолонка 'Age' не знайдена.")

# Розгляньте можливість видалення або кодування пропущених значень для Cabin.
# Видаляємо колонку Cabin через велику кількість пропусків
if 'Cabin' in titanic.columns:
    titanic.drop('Cabin', axis=1, inplace=True)
    print("Колонка Cabin видалена.")
else:
    print("Колонка Cabin не знайдена.")

# b. Перетворіть категоріальні ознаки (наприклад, Sex, Embarked) у числові значення,
# використовуючи one-hot encoding або label encoding.
# Use the actual column names identified earlier
categorical_cols = ['Sex', 'Embarked']
categorical_cols = [col for col in categorical_cols if col in titanic.columns]  # Keep only the columns that are actually present
if categorical_cols:
    titanic = pd.get_dummies(titanic, columns=categorical_cols, drop_first=True)
    print("\nКатегорійні ознаки", categorical_cols, "перетворені на числові за допомогою one-hot encoding.")
else:
    print("\nЖодна з категорійних ознак Sex та Embarked не знайдена.")

# c. Перевірте можливість створення нових ознак:
# Виділення титулу з поля Name.
if 'Name' in titanic.columns:
    titanic['Title'] = titanic['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
    print("\nВиділено титули з колонки Name. Ось перші 5 унікальних титулів:")
    if 'Title' in titanic.columns:
        print(titanic['Title'].unique()[:5])
    else:
        print("Колонка Title не знайдена")
else:
    print("Колонка Name не знайдена")

# Обчислення розміру родини на борту (сума SibSp та Parch).
if 'SibSp' in titanic.columns and 'Parch' in titanic.columns:
    titanic['FamilySize'] = titanic['SibSp'] + titanic['Parch'] + 1
    print("\nСтворено нову ознаку FamilySize.")
else:
    print("\nКолонки SibSp або Parch не знайдені. Неможливо створити ознаку FamilySize.")

print("\nПерші 5 рядків обробленого набору даних:")
print(titanic.head())
print("\nРозмір обробленого набору даних (рядки, стовпці):", titanic.shape)
```

```python
# Візуалізація даних за допомогою Matplotlib та Seaborn Кроки
# 4.1 Для набору Iris
import matplotlib.pyplot as plt
import seaborn as sns

# Завантаження набору даних Iris
iris = sns.load_dataset("iris")

# a. Побудуйте гістограми розподілу для кожної числової ознаки
plt.figure(figsize=(12, 8))  # Збільшимо розмір плоту для кращої читабельності

plt.subplot(2, 2, 1)  # Розміщуємо 4 графіки на одній площині (2 рядки, 2 стовпці, 1-й графік)
plt.hist(iris['sepal_length'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Sepal Length')

plt.subplot(2, 2, 2)  # 2-й графік
plt.hist(iris['sepal_width'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Sepal Width')

plt.subplot(2, 2, 3)  # 3-й графік
plt.hist(iris['petal_length'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Petal Length')

plt.subplot(2, 2, 4)  # 4-й графік
plt.hist(iris['petal_width'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Petal Width')

plt.tight_layout()  # Автоматичне коригування відступів для запобігання перекриттю
plt.show()


# b. Побудуйте накладені гістограми для порівняння розподілів однієї ознаки для різних видів ірисів.
plt.figure(figsize=(8, 6))
species = iris['species'].unique()
colors = ['red', 'green', 'blue']

for sp, col in zip(species, colors):
    subset = iris[iris['species'] == sp]
    plt.hist(subset['sepal_length'], bins=15, alpha=0.5, label=sp, color=col, edgecolor='black')

plt.title('Sepal Length Histogram by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# c. Побудуйте розділені гістограми для детального аналізу розподілів по групах.
g = sns.FacetGrid(iris, col="species", height=4, aspect=1)
g.map(plt.hist, "sepal_length", bins=15, color='coral', edgecolor='black')
plt.show()

# d. Побудуйте гістограму з накладеною кривою оцінки щільності розподілу (KDE).
plt.figure(figsize=(8, 6))
sns.histplot(iris['petal_length'], bins=15, kde=True, color='purple', edgecolor='black')
plt.title('Histogram and KDE for Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()

# e. Побудуйте кумулятивну гістограму для аналізу накопичуваних частот.
plt.figure(figsize=(8, 6))
plt.hist(iris['sepal_width'], bins=15, cumulative=True, color='teal', edgecolor='black')
plt.title('Cumulative Histogram for Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Cumulative Frequency')
plt.show()

# f. Створіть парну діаграму (pairplot) для перегляду взаємозв’язків між ознаками з розподілом за класами.
sns.pairplot(iris, hue='species', markers=["o", "s", "D"])
plt.show()
```

```python
#4.2. Для набору Titanic

import matplotlib.pyplot as plt
import seaborn as sns

# Завантаження набору даних Titanic
titanic = sns.load_dataset('titanic')

# a. Побудуйте парний графік (pairplot) для набору Titanic
sns.pairplot(titanic.dropna(), hue='survived', vars=['age', 'fare'])
plt.suptitle('Парний графік ознак для Titanic', y=1.02)
plt.show()

# b. Створіть діаграму розподілу кількості пасажирів, що вижили та не вижили.
sns.countplot(x='survived', data=titanic)
plt.title('Розподіл виживання')
plt.show()

# c. Побудуйте графіки залежності виживання від ключових ознак:
# Стать (Sex) – виживання за статтю
# Клас пасажира (Pclass) – виживання за класом

# Виживання за статтю
sns.countplot(x='sex', hue='survived', data=titanic)
plt.title('Виживання за статтю')
plt.show()

# Виживання за класом
sns.countplot(x='pclass', hue='survived', data=titanic)
plt.title('Виживання за класом')
plt.show()

# d. Побудуйте гістограми для числових змінних Age і Fare, щоб дослідити їх розподіл.
plt.figure(figsize=(12, 5))  # Збільшимо розмір плоту для розміщення двох графіків поруч

plt.subplot(1, 2, 1)  # Розміщуємо 2 графіки в одному рядку, 2 стовпці, 1-й графік
titanic['age'].hist(bins=30)
plt.title('Розподіл віку')
plt.xlabel('Вік')
plt.ylabel('Кількість пасажирів')

plt.subplot(1, 2, 2)  # 2-й графік
titanic['fare'].hist(bins=30)
plt.title('Розподіл вартості проїзду')
plt.xlabel('Вартість проїзду')
plt.ylabel('Кількість пасажирів')

plt.tight_layout()  # Автоматичне коригування відступів
plt.show()
```

```python
# 5 Ручне розділення даних на тренувальну і тестову

import numpy as np
import pandas as pd
import seaborn as sns

# Завантаження набору даних Titanic
titanic = sns.load_dataset('titanic')

# 1. Отримання даних та цільової змінної
data = titanic.drop('survived', axis=1)  # Видаляємо 'survived', щоб отримати лише ознаки
target = titanic['survived']  # 'survived' - наша цільова змінна

# 2. Ініціалізація генератора випадкових чисел для відтворюваності результатів
np.random.seed(42)

# 3. Створення масиву індексів
indices = np.arange(data.shape[0])  # Створюємо масив індексів від 0 до кількості зразків

# 4. Перемішування індексів
np.random.shuffle(indices)  # Перемішуємо індекси випадковим чином

# 5. Визначення розміру тренувальної вибірки (70% від загальної кількості)
train_size = int(0.7 * len(indices))

# 6. Розділення індексів на тренувальні та тестові
train_indices = indices[:train_size]  # Перші 70% індексів для тренувальної вибірки
test_indices = indices[train_size:]  # Решта 30% індексів для тестової вибірки

# 7. Створення тренувальних та тестових наборів даних на основі розділених індексів
X_train = data.iloc[train_indices]  # Використовуємо .iloc для вилучення рядків за індексами
y_train = target.iloc[train_indices]
X_test = data.iloc[test_indices]
y_test = target.iloc[test_indices]

# 8. Виведення розмірів тренувального та тестового наборів для перевірки
print("Розмір тренувального набору (X_train):", X_train.shape)
print("Розмір тренувального набору (y_train):", y_train.shape)
print("Розмір тестового набору (X_test):", X_test.shape)
print("Розмір тестового набору (y_test):", y_test.shape)
```
